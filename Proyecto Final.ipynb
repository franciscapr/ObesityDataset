- Se espera idenificar entre los nuevos pacientes aquellos que tienen antecedentes familiares de sobrepeso y aquellos que no los tienen. Para lograr esto, utilizare el algoritmos de los K-veecinos más cercanos.

```python
# Filtramos los datos del df data que presentan antecedentes familiares de sobrepeso y aquellos que no
con_antecedentes = data[data["Antecedentes_familiares_sobrepeso"]==1]
sin_antecedentes = data[data["Antecedentes_familiares_sobrepeso"]==0]

# Utilizamos el método scatter para hacer un gráfico de dispersi
plt.scatter(con_antecedentes["Edad"], con_antecedentes["IMC"],
            marker="*", s=50, color="skyblue",
            label="Con Antecedentes Sobrepeso")

plt.scatter(sin_antecedentes["Edad"], sin_antecedentes["IMC"],
            marker="*", s=50, color="orange",
            label="Sin Antecedentes Sobrepeso")

plt.ylabel("Indice masa corporal")
plt.xlabel("Edad")
plt.legend(bbox_to_anchor=(1, 0.2))
plt.show()  
```

### Feature Selection

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
# Asiganamos las variables a utilizar como features y target
X = data[["Edad", "IMC"]]
```

```python
X.head(2)
```

```python
y = data["Antecedentes_familiares_sobrepeso"]
```

```python
y.head(2)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

## Standard Features

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Train

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

## Predict

```python
y_pred = knn.predict(X_test)
```

```python
y_pred
```

## Accuracy

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

## Visualization

```python
edad = 29
imc = 237000

# Escalar los datos del nuevo paciente
paciente = scaler.transform([[edad, imc]])
print("Clase", knn.predict(paciente))
print("Probabilidad", knn.predict_proba(paciente))
# Calcular la clase y probabilidades

# Codigo para graficas

plt.scatter(con_antecedentes["Edad"], con_antecedentes["IMC"],
            marker="*", s=100, color="skyblue", label="Con Antecedentes Clase 1")
plt.scatter(sin_antecedentes["Edad"], sin_antecedentes["IMC"],
            marker="*", s=100, color="orange", label="Sin Antecedentes Clase 0")

plt.scatter(edad, imc, marker="P", s=100, color="red", label="Paciente")
plt.ylabel("Indice masa corporal")
plt.xlabel("Edad")
plt.legend(bbox_to_anchor=(1, 0.3))
plt.show()
```
