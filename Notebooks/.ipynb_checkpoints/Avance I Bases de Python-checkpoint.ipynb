## 1. `CitibikeTripdata.csv`
### Descripción:

- El archivo `CitibikeTripdata.csv` contiene información detallada sobre los viajes realizados utilizando el sistema de bicicletas compartidas conocido como Citibike. Este sistema, operado por Motivate en colaboración con Citibank, ofrece servicios de alquiler de bicicletas en la ciudad de Nueva York, permitiendo a los usuarios rentar bicicletas por períodos cortos, especialmente para desplazamientos dentro de la ciudad.

- El conjunto de datos proporciona una amplia gama de información relacionada con los viajes, incluyendo la duración de los mismos, la ubicación de inicio y fin en las estaciones designadas, así como la fecha y hora tanto de inicio como de finalización de cada viaje. Además, se incluye la identificación única de cada bicicleta utilizada en el sistema.

- Es importante destacar que estos datos son una valiosa fuente de información para la investigación en áreas como la movilidad urbana, la planificación del transporte y el análisis de patrones de desplazamiento en entornos urbanos. Este conjunto de datos específico fue extraído del libro "Practical Python Data Wrangling and Data Quality".

### Variables releventes

- `tripduration`: La duración del viaje es una medida fundamental que puede proporcionar información sobre el comportamiento de los usuarios y la demanda del servicio de bicicletas compartidas.

- `starttime` y `stoptime`: Estas variables indican el momento en que comenzó y terminó cada viaje, lo que permite analizar patrones de uso a lo largo del tiempo, como horas punta y períodos de baja demanda.

- `start station id` y `end station id`: Identificadores de las estaciones de inicio y fin de cada viaje. Estas variables son cruciales para analizar la distribución geográfica de los viajes y la utilización de las estaciones de bicicletas.

- `usertype`: Esta variable indica el tipo de usuario, que podría ser "suscriptor" o "cliente ocasional". Esto podría ser útil para segmentar el análisis según el tipo de usuario y comprender mejor sus comportamientos y necesidades.

- `gender` y `birth year`: Estas variables pueden proporcionar información demográfica sobre los usuarios, lo que podría ser relevante para entender mejor quiénes están utilizando el servicio de bicicletas compartidas y cómo se distribuye su uso entre diferentes grupos de edad y género.

## 2. `loan_data.csv`
### Descripción: 
El archivo `loan_data.csv` contiene una recopilación de datos relacionados con préstamos bancarios. Es un conjunto de datos ampliamente utilizado en libros, análisis financieros, prácticas de ciencia de datos, análisis de datos, aprendizaje automático, modelado predictivo, entre otros campos.- 

Dentro de este conjunto de datos, se incluyen diversos atributos que proporcionan información detallada sobre cada préstamo. Esto incluye identificadores únicos para los préstamos, monto del préstamo, tasas de interés asociadas, plazos de pago, estados del préstamo (por ejemplo, pagado, atrasado), tipo de préstamo, historial de pagos, información sobre garantías y otros detalles relevantes sobre los préstamos bancario- s.

La utilidad de estos datos radica en su capacidad para respaldar análisis financieros, evaluación de riesgos crediticios, elaboración de modelos predictivos y otros estudios relacionados con la gestión de préstamos y la toma de decisiones financie- ras.

Es importante señalar que este conjunto de datos en particular fue extraído del libro "Estadísticas Prácticas Para Ciencia de Datos con R y Python".

### Variables releventes

- `status`: El estado del préstamo (pagado, atrasado, etc.) es fundamental para comprender el resultado de los préstamos y su impacto en la gestión de riesgos crediticios.

- `loan_amnt`: El monto del préstamo es importante para comprender la cantidad de dinero prestado y su relación con la capacidad de pago del prestatario.

- `term`: El plazo del préstamo puede influir en la capacidad de pago mensual del prestatario y en la probabilidad de incumplimiento.

- `annual_inc`: Los ingresos anuales del prestatario son críticos para evaluar su capacidad de pago y determinar el nivel de riesgo crediticio.

- `dti`: La relación deuda-ingresos (DTI) es una medida clave de la capacidad del prestatario para manejar nuevas deudas en relación con sus ingresos actuales.

- `purpose`: El propósito del préstamo puede proporcionar información sobre la finalidad del préstamo y su impacto en la capacidad de pago del prestatario.

- `revol_bal`: El saldo revolvente (revolving balance) indica la cantidad de deuda que el prestatario mantiene en cuentas de crédito renovables, lo que puede afectar su capacidad de pago y su salud financiera general.

- `revol_util`: La utilización del crédito revolvente (revolving credit utilization) es otra medida importante de la salud financiera del prestatario y su capacidad de pago.

- `delinq_2yrs_zero`: Indica si el prestatario ha tenido retrasos en los pagos en los últimos 2 años, lo que puede ser indicativo de su historial de crédito y su capacidad de pago.

- `open_acc`: El número de cuentas abiertas puede proporcionar información sobre la diversificación del crédito del prestatario y su capacidad de pago.

- `grade`: La calificación del préstamo (A, B, C, etc.) proporciona una evaluación del riesgo crediticio asociado al préstamo.

- `outcome`: El resultado del préstamo (aprobado, denegado, etc.) es crucial para comprender los resultados del proceso de solicitud de préstamo.

## 3. `ObesityDataSet.csv`

### Descripción:

El conjunto de datos proporciona información crucial para estimar los niveles de obesidad en individuos de México, Perú y Colombia, fundamentándose en sus hábitos alimenticios y estado físico. Compuesto por 17 atributos y 2111 registros, cada entrada está marcada con la variable de clase "NObesity", permitiendo una variedad de análisis como clasificación, agrupación y regresión. Estos análisis se basan en categorías que incluyen Peso Insuficiente, Peso Normal, Sobrepeso Nivel I y II, así como tres tipos de Obesidad: Tipo I, Tipo II y Tipo III. Esta diversidad de categorías facilita una comprensión profunda y detallada de los patrones de obesidad en las poblaciones de los tres países mencionados. Los datos presentan una oportunidad valiosa para investigadores y profesionales de la salud, permitiendo la identificación de factores de riesgo, la evaluación de la eficacia de intervenciones preventivas y la formulación de políticas de salud pública dirigidas a abordar el problema de la obesidad en la región latinoamericana. Este conjunto de datos es una herramienta invaluable para avanzar en la comprensión y el tratamiento de la obesidad, contribuyendo así a la promoción de estilos de vida saludables y la mejora del bienestar de la población.

### Variables releventes

- `Gender` --- > Género: Indica el género del individuo.
- `Age` ---> Edad: Representa la edad del individuo.
- `Height` ---> Altura: Mide la altura del individuo en metros.
- `Weigh` ---> Peso: Indica el peso del individuo en kilogramos.
- `family_history_with_overweight` ---> Antecedentes familiares de sobrepeso: Refleja si el individuo tiene antecedentes familiares de sobrepeso.
- `FAVC` ---> Consumo frecuente de alimentos altos en calorías: Indica si el individuo consume regularmente alimentos altos en calorías.
- `FCVC` ---> Frecuencia de consumo de verduras: Representa la frecuencia con la que el individuo consume verduras.
- `NCP` ---> Número de comidas principales: Indica el número de comidas principales que el individuo consume al día.
- `CAEC` ---> Consumo de alimentos entre comidas: Refleja el consumo de alimentos entre comidas del individuo.
- `SMOKE` ---> Fumador o no: Indica si el individuo es fumador o no.
- `CH2O` ---> Consumo diario de agua: Mide el consumo diario de agua del individuo en litros.
- `SCC` ---> Monitoreo del consumo de calorías: Indica si el individuo monitorea su consumo de calorías.
- `FAF` ---> Frecuencia de actividad física: Representa la frecuencia con la que el individuo realiza actividad física.
- `TUE` ---> Tiempo utilizando dispositivos tecnológicos: Indica el tiempo que el individuo pasa utilizando dispositivos tecnológicos.
- `CALC` ---> Consumo de alcohol: Refleja el consumo de alcohol del individuo.
- `MTRANS` ---> Medio de transporte utilizado: Indica el medio de transporte que el individuo utiliza.
- `NObeyesdad` ---> Nivel de obesidad deducido: Representa el nivel de obesidad deducido del individuo.
