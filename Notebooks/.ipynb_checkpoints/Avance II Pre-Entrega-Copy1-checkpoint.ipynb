- Pasamos como argumento color para proporcionar el valor de skyblue.
- Asiganmos edgecolor con un valor de white para el borde de las barras del histograma, esto nos proporciona mejor division entre las barras, diferencia.
- Asiganmos un alpha de 0.8 es decir un grado de opacidad.
- Aignamos los grid con valor True para mostrar las grillas del grafico.

```python
# Definimos el número del bins para le histograma
min_peso = min(df["Peso"])
max_peso = max(df["Peso"])
n_bins = int(np.ceil((max_peso - min_peso) / 5))  # Ajustar el ancho de los bins según tu preferencia

# Definimos el tamaño del linzo 8 * 3
plt.figure(figsize=(8, 3))
# Llamamos al métod hist para el histrograma, pasamos la variable peso y los argumentos bins que contiene la cantidad de
# bins definidos anteriormente, argumentamos el color con el valor de skyblue y el margen con white, y una opacidad de 0.8
plt.hist(df["Peso"], bins=n_bins, color='skyblue', edgecolor="white", alpha=0.8)

# Definimos el titulo del gráfico
plt.title('Histograma de Pesos')
# Definimos la etiqueta del eje x
plt.xlabel('Peso')
# Definimos la etiqueta del eje y
plt.ylabel('Frecuencia')

# Definimos grillas para el gráfico
plt.grid(True)
# Imprimimos el gráfico
plt.show()
```

# Pie Chart

- Utilizamos el Pie chaart para visualizar la frecuencia en cada categoria de peso.
- Este gráfico muetras los valores de pesos agrupados en los intervalos "bins".
- Cada segmento del gráfico contiene la cantidad de observaciones que caen dentro de ese intervalo de ese rango de pesos.
- Definimos un lienzo de tamaño 6 * 3.

```python
# Definimos los grupos de pesos para el gráfico de pastel
bins = [0, 50, 100, 150, 200] 

# Creamos las categorias para los grupos de pesos para el gráfico
categorias = ['0-50', '51-100', '101-150', '151-200']

# Agrupamos los pesos, creamos una variable Grupo_pesos en donde le asignamos los pesos, pasando como arguementos
# los bins que definimos y las etiquetas que seran las categorias del gráfico
df['Grupo_Peso'] = pd.cut(df['Peso'], bins=bins, labels=categorias)

# Asignamos a la variable frecuencia por grupos los grupos de peso calculados
frecuencia_por_grupo = df['Grupo_Peso'].value_counts()

# Obtenemos las etiquetas de los grupos de pesos y las frecuencias
categorias = frecuencia_por_grupo.index
frecuencias = frecuencia_por_grupo.values

# Calculamos el porcentaje de cada grupo 
porcentajes = frecuencias / frecuencias.sum() * 100

# Definimos el tamaño del lienzo
plt.figure(figsize=(6, 6))
# Definimos el gráfico de pie, pasamos argumentos como porcentajes que es la variable calculada, las etiquetas que son
# las categorias, autopct que nos prorporciona formato de punto flotante "float", startangle nos proporciona el angulo de 
# inicio del gráfico en este caso comienza desde el angulo 140 grados, labeldistance nos permite controlar la distancia de las
# etiquetas.
plt.pie(porcentajes, labels=categorias, autopct='%1.1f%%', startangle=140, labeldistance=1.1)
# Titulo del gráfico
plt.title('Distribución de la muestra por grupos de Peso')
# Esto asegura que el gráfico sea un círculo en lugar de una elipse
plt.axis('equal') 
# Imprimimos el gráfico
plt.show()
```

### Interpretación

- El gráfico de pastel nos brinda una visión de la distribución de la muestra según grupos de peso predefinidos.
- Los grupos de peso están divididos en rangos de 0-50, 51-100, 101-150 y 151-200 kilogramos.
- El segmento con mayor representación en la muestra corresponde al grupo de peso entre 51 y 100 kilogramos, abarcando el 57.5% de todas las observaciones.
- El segundo grupo más representativo es el de 101-150 kilogramos, comprendiendo el 32.9% de las observaciones.
- Por otro lado, los grupos de peso de 0 a 50 kilogramos y de 151 a 200 kilogramos presentan una menor participación, representando solo el 8.6% y el 1% de las observaciones, respectivamente.

## 3. ¿ Las personas con antecedentes familiares de sobrepeso, tienden a tener obesidad ?

- Hipótesis: Se espera que según los antecendentes familiares de sobrepeso u obesidad, se tenga una nivel de obesidad.

# Pie Chart

- Utilizamos un gráfico de pastel para ver la proporción de los grados de sobrepeso y obesidad de las observaciones.

```python
# Creamos una variable a la que le pasamos la cantidad de categorias
frecuencia_por_categoria = df['Nivel_obesidad'].value_counts()

# Creamos dos variables con las frecuencias como valores y con las categorias como indeces
categorias = frecuencia_por_categoria.index
frecuencias = frecuencia_por_categoria.values

# Calculamos el porcentaje de cada una de las categorias
porcentajes = frecuencias / frecuencias.sum() * 100

# Creamos el lienzo de tamaño 5 * 5
plt.figure(figsize=(5, 5))
# Llamamos al métodos pie para le grafico de torta, le pasamos como argumentos la variable porcentajes, las etiquetas que son
# las ategorias, el argumetno autopct para que nos de una valor en punto flotante "float"
plt.pie(porcentajes, labels=categorias, autopct='%1.1f%%', startangle=140)
# Definimos el titulo del gráfico
plt.title('Distribución de Categorías')

plt.axis('equal') 
plt.show()
```

### Interpretación

- En nuestro gráfico, observamos que la obesidad de tipo I es el grupo con mayor proporción de observaciones, representando un 16.6% del total. Le sigue la obesidad de tipo III, con un 15.3% de las observaciones.
- El peso normal constituye aproximadamente el 13.6% de las observaciones, mostrando una distribución significativa.
- En contraste, el bajo peso presenta la menor proporción de observaciones, con un 12.9%, lo que indica que es el grupo menos representado en nuestro conjunto de datos.

# Bar plot

- El gráfico de barras nos proporciona una visualización de la distribución de las observaciones en función de los antecedentes familiares de sobrepeso u obesidad. Este análisis busca examinar la relación entre los antecedentes familiares de sobrepeso y obesidad con el nivel de obesidad presente en las observaciones.

```python
# Generamos las variables que utilizaremos, para esto realizamos filtros para las observaciones con antecedentes
# y para las observaciones sin antecedentes de sobrepeso
con_antecedentes = df[df['Antecedentes_familiares_sobrepeso'] == 'yes']
sin_antecedentes = df[df['Antecedentes_familiares_sobrepeso'] == 'no']


# Realizamos el conteo de la cantidad de observaciones con antecedentes y sin antecedentes
con_antecedentes_counts = con_antecedentes['Nivel_obesidad'].value_counts()
sin_antecedentes_counts = sin_antecedentes['Nivel_obesidad'].value_counts()

# Creamos un linzo de tamaño 5 * 5
plt.figure(figsize=(5, 5))
# Llamamos al método bar para el gráfico de barras y le pasamos los indices y los valores, algunos arugmentos como la opacidad
# de los datos, la etiqueta si tiene antecentes o no tiene antecedentes y el color especificos para cada etiqueta
plt.bar(con_antecedentes_counts.index, con_antecedentes_counts.values, alpha=0.4, label='Con Antecedentes', color="red")
plt.bar(sin_antecedentes_counts.index, sin_antecedentes_counts.values, alpha=0.7, label='Sin Antecedentes', color="skyblue")

# Definimos el titulo del gráfico
plt.title('Distribución de Nivel de Obesidad por Antecedentes Familiares de Sobrepeso')
# Definimos la etiqueta del eje x 
plt.xlabel('Nivel de Obesidad')
# Definimos la etiqueta del eje y
plt.ylabel('Cantidad')
# Definimos la rotación "posición" que tendran la eiquetas para que no se solapen, le pasamos 45 grados
plt.xticks(rotation=45, ha='right')

# Añadimos una leyenda para las dos clasificaciones, sin antecedentes y con antecedentes
plt.legend()
# Imprimimos el gráfico
plt.show()
```

### Interpretación

- En el eje x se clasifican las categorias de nivel de Obesidad.
- En el eje y se distribuye la frecuencia según cada una de las categorias.
- Nuestro grafico clasifica los datos con antecedentes y sin antecedentes de obsedidad familiar.
- Como se observa en todos los niveles de obseidad desde la I hasta la III y sobrepeso tenemos personas con familiares con antecedentes de obesidad.

# Boxplot

- Realizamos un boxplot entre la variable peso y los clasificamos en genero (Female y Male).

```python
# Generamos las variables a utilizar, en este caso genero y peso
genero = df["Género"]
peso = df["Peso"]

# Definimos el tamaño del lienoz
plt.figure(figsize=(5, 3))

# Creamos el boxplot con los argumentos a utilizar y las variables
sns.boxplot(data=df, x=genero, y=peso, notch=True, flierprops=dict(markerfacecolor="orange"),
            showmeans=True, meanline=True, boxprops=dict(facecolor="skyblue"),
            whiskerprops=dict(color="black", linewidth=1), capprops=dict(color="blue", linewidth=2))

# Definimos el titulo
plt.title("Boxplot de Peso por Género")
# Definismo la etiqueta el eje x
plt.xlabel("Género")
# Definimos la etiqueta del eje y
plt.ylabel("Peso")

# Imprimimos el gráfico
plt.show()
```

### Interpretación

- Podemos observar que el peso máximo registrado para hombres es aproximadamente 130 kg, mientras que para mujeres es alrededor de 170 kg. Esto sugiere que en general, las mujeres tienden a tener pesos más altos en comparación con los hombres.
- La mediana del peso para las mujeres está cerca de 80 kg, mientras que para los hombres está alrededor de 90 kg. Esto indica que la mitad de las mujeres tienen un peso igual o inferior a 80 kg, y la mitad de los hombres tienen un peso igual o inferior a 90 kg.
- También podemos notar la presencia de un valor atípico en el grupo de hombres, que supera los 160 kg. Este valor atípico indica la presencia de un peso extremadamente alto en el conjunto de datos de hombres.
